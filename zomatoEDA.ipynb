```python
df['rate'] = df['rate'].apply(lambda x: str(x).replace('/5',''))
```

```python
def rate(x):
    if x=='NEW' or x=='-':
        return np.nan
    else:
        return float(x)
    
df['rate'] = df['rate'].apply(rate)
```

```python
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].replace(np.nan,df["approx_cost(for two people)"].mode()[0]).map(lambda x :  float(str(x).replace(",","")))
```

```python
df['rate'] = df['rate'].fillna(df['rate'].median())
df['location'] = df['location'].fillna(df['location'].mode()[0])
df['rest_type'] = df['rest_type'].fillna(df['rest_type'].mode()[0])
df['cuisines'] = df['cuisines'].fillna(df['cuisines'].mode()[0])
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
```

```python
extra = df['rest_type'].value_counts()[df['rest_type'].value_counts()<1000]
def treat_rest(x):
    if x in extra:
        return 'others'
    else:
        return x
df['rest_type'] = df['rest_type'].map(treat_rest)
```

```python
others = df['cuisines'].value_counts()[df['cuisines'].value_counts()<100]
def cuisine(x):
    if x in others:
        return 'others'
    else:
        return x
df['cuisines'] = df['cuisines'].map(cuisine)
```

```python
"""def loc(x):
    if str(x)[:5].lower()=="koram":
        return "Koramangala"
    else:
        return x
df["location"]=df["location"].map(loc)"""
location = df['location'].value_counts()[df['location'].value_counts()<300]
def locations(x):
    if x in location:
        return 'others'
    else:
        return x
df['location'] = df['location'].apply(locations)
```
